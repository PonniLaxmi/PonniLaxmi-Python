# 5. Aggregating data with transform

```python
# Creating loyalty flag
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# deriving columns with loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# printing the freq of column loyality_flag
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
# Q6) Different type of spenders
ords_prods_merge['average_price']= ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#derive column conditions
ords_prods_merge.loc[ords_prods_merge['average_price']<10, 'spending_flag']= 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price']>=10, 'spending_flag']= 'High spender'
```

```python
# Q7 create order frequency flag
ords_prods_merge['order_frequency_flag']= ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
#derive column conditions
ords_prods_merge.loc[ords_prods_merge['order_frequency_flag']>20, 'customer_frequency_flag']= 'Non frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency_flag']>10) & (ords_prods_merge['order_frequency_flag']<=20)  , 'customer_frequency_flag']= 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency_flag']<=10, 'customer_frequency_flag']= 'frequent customer'
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'spending_flag', 'customer_frequency_flag']].head(60)
```
