#### There are no duplicates found and there are 11259 missing values in the first_name column. Hence, it can't be deleted as it might affect the analysis. We can leave the missing values as such now.

```python
# 2. Addressing missing data types
for col in df_customer.columns.tolist():
  weird = (df_customer[[col]].applymap(type) != df_customer[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customer[weird]) > 0:
    print (col)
```

```python
# changing the column first_name data type
df_customer['first_name'] = df_customer['first_name'].astype('str')
```

```python
# checking the column dtype
df_customer['first_name'].dtype
```

```python
# checking again for mixed data type 
for col in df_customer.columns.tolist():
  weird = (df_customer[[col]].applymap(type) != df_customer[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customer[weird]) > 0:
    print (col)
```
